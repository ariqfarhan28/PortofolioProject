# **Membandingkan 2 kelompok**






```python
import pandas as pd
pizzas = pd.read_csv('pizzas.csv')
```

```python
pizzas.head()
```

# Uji Normalitas 2 kelompok

```python
unit_1 = pizzas['Making Unit 1']
unit_2 = pizzas['Making Unit 2']
```

```python
H0 = 'Data distribution is normal'
Ha = 'Data distribution not normal'

alpha = 0.05

from scipy.stats import shapiro

#create class for data check
def datacheck(df):
  for columnName, columnData in pizzas.items():
    print("Shapiro Test Results of '{}' ".format(columnName))
    p_value = round(shapiro(columnData.values)[1], 2)
    if p_value>alpha:
      print(f"{p_value} > {alpha}, Because P-value > alpha. then '{columnName}' {H0}")
    else:
      print(f"{p_value} <= {alpha}, Because P-value <= alpha. then '{columnName}' {Ha}")
#call class
datacheck(pizzas)
```

# Uji Variansi 2 kelompok

```python
H0 = 'Unit 1 and Unit 2 have the same variance '
Ha = 'unfortunately unit 1 and unit 2 do not have the same variance'

from scipy.stats import levene

def check_variances(df):
  print("Variances Test Results")
  p_value = round(levene(unit_1, unit_2)[1],2)
  if p_value>alpha:
    print(f"{p_value} > {alpha}. {H0}")
  else:
    print(f"{p_value} <= {alpha}. {Ha}")

check_variances(pizzas)
```

# 2 sample t-test for equal variance

```python
H0 = 'There is no significant difference in unit 1 and unit 2'
Ha = 'There a significant difference in unit 1 and unit 2'

from scipy.stats import ttest_ind

def t_test(df):

  print( "Sample T Test Results")
  test_results = ttest_ind(unit_1, unit_2, equal_var=True)
  p = round(test_results[1],2)

  if p>alpha:
    print(f"{p} > {alpha}. {H0}")
  else:
    print(f"{p} <= {alpha}. {Ha}")

t_test(pizzas)
```
